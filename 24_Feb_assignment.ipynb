````python
### ANS 2.

import pandas as pd

#exampe datatframe
data = { "A": [10,20,30], "B": [40,50,60] , "C" : [70,80,90]}
df = pd.DataFrame(data)

# function for reindexing the dataframe as per our requirements
def reindexed_dataframe(df):
  new_index = pd.RangeIndex(start =1 , stop = len(df)*2 , step = 2)
  df =  df.reset_index(drop =True)
  df.index =  new_index
  return df

# resetting index using the function reindexed_dataframe
df_reindexed = reindexed_dataframe(df)
print(df_reindexed)

print("-----------------------------------------------------------------------------")

# another simple way can be there as well without making a function , all those values with no corresponding index get occupied with NaN values
df = pd.DataFrame(data)
print(df.reindex([1,3,5]))

````

````python
##ANS 4.

# example DataFrame:
tools = ["EC2", "bean stalk" , "docker", "lambda" , "jenkins" , "CI/CD"  , "linux"]
df = pd.DataFrame(tools , columns = ["tools"])

# function to count the length and append the same into the list
def counting(df):

  l= []
  for  i in range(0,len(df)):
    l.append(len(df['tools'][i]))


  # print(l)

  df['word_count']= l
  return df

# function called
counting(df)
````

````python
# ANS 7.

emails = ["abc@gmail.com","suraj@yahoo.com", "test@example.com","pista@gmail.com"]

df= pd.DataFrame(emails, columns =['emails'])

def splitting(df):
  k = []
  for i in range(len(df)):
    k.append(df['emails'][i].split("@")[0])

  df['user_name'] = k
  return df

# print(k)

splitting(df)
````

````python
#ANS 10.
data = {"sales" : [4500,5500,6000,1200,7800], "date" : ['1july' ,'7july','14july','21juy','28july' ] }
df = pd.DataFrame(data,columns = ['sales','date'])
df

def calculate_moving_average(df):
    window_size = 7
    df['MovingAverage'] = df['sales'].rolling(window=window_size, min_periods=1).mean()
    return df

calculate_moving_average(df)
````

````python

# ANS 11.
import pandas as pd

# Example usage
data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']
}

#function
def add_weekday_column(df):
    # Convert the 'Date' column to a datetime type
    df['Date'] = pd.to_datetime(df['Date'])

    # Add a new column 'Weekday' containing the weekday names
    df['Weekday'] = df['Date'].dt.strftime('%A')

    return df



df = pd.DataFrame(data)
df_with_weekday = add_weekday_column(df)
df_with_weekday
````

````python
# ANS 12.

#example dataframe
data = {'Date': ['2023-01-01', '2023-01-15', '2023-01-31', '2023-02-01', '2023-02-15']}
df  = pd.DataFrame(data)

#function
def selected_rows_bw_dates(df):
  df['Date'] =  pd.to_datetime(df['Date'])

  start_date = pd.to_datetime('2023-01-01')
  end_date = pd.to_datetime('2023-01-31')

  selected_rows = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
  return selected_rows

#function call
selected_rows_bw_dates(df)


````
