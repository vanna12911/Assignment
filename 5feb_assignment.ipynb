In object-oriented programming, a class is a blueprint or template that defines the properties and behaviors of a particular type of object. An object, on the other hand, is an instance of a class - it is a specific entity that can be created based on the class definition.

In simpler terms, a class can be thought of as a recipe or a set of instructions for creating objects, while an object is the actual dish that is created using that recipe.

```python

# Define the Car class
class cars:
  
  # Constructor method - initializes the object with make, year,model and color
  def __init__(self, make, year,model,color ):
    self.make = make
    self.year= year
    self.model = model
    self.color= color

# Method to start the car
  def start(self):
    print(f"{self.make} {self.model} has been started")


# Method to accelerate the car
  def accelerate(self):
    print(f"{self.make} {self.model} has been accelerated")


# Method to stop the car
  def stop(self):
    print(f"{self.make} {self.model} has been stopped")


# Create an instance of the Car class
my_car =cars("Mahindra" ,2021 ,"Thar","black")

# call methods on my_car object
my_car.start()
my_car.accelerate()
my_car.stop()
```

In object-oriented programming, the \_\_init\_\_() function (or method) is a special method that is called when an object of a class is created. It is used to initialize the object's attributes (or properties) with values provided by the user or with default values defined in the class.

Here's an example to illustrate why the __init__() function is used:

code without a constructor, here Without the __init__() method, we would have to manually set the attributes of the object after creating it:

```python

class Person:

  name = " "
  age = 0    
    
    # def __init__(self, name, age):
    #     self.name = name
    #     self.age = age
        
    
  def print_info(self):
      print(f"Name: {self.name}, Age: {self.age}")

p1 = Person()
p1.name = "pista"
p1.age =20
p1.print_info()
```

This approach is more error-prone and less convenient than using the __init__() method to initialize the attributes at the time of object creation.

here's a piece of code with constructor:
```python

class Person:   
  
  def __init__(self, name, age):
      self.name = name
      self.age = age
  
  def print_info(self):
      print(f"Name: {self.name}, Age: {self.age}")


p1 = Person("Pista",20)
p1.print_info()
```
In summary, the __init__() method is used to initialize the attributes of an object with values provided by the user or with default values defined in the class. It is a convenient and error-free way of creating objects with predefined properties.


####ANS(4):
In object-oriented programming, self is a reference to the object that is calling the method. It is used to access the attributes and methods of the object within the method.

When a method is called on an object, Python automatically passes the object itself as the first argument to the method, which is usually referred to as self. The self reference allows the method to access the object's attributes and methods, which can be used to modify the object's state or behavior.

In short , the self reference is used in object-oriented programming to access and modify the attributes and methods of an object within a method. It is a way to reference the specific instance of the class that is calling the method amd self is not a keyword in python therefore any other valid name can be used in place of self except the predefined keywords from python. 


####ANS(5):
Inheritance is a fundamental concept in object-oriented programming that allows a class to inherit properties and behaviors from another class. The class that inherits from another class is called a derived class or subclass, and the class that is inherited from is called a base class or superclass.

In Python, there are mainly five types of inheritance:

1.<b><u> Single level inheritance :</b></u>   When a subclass inherites properties and behaviors from a single parent/base class.

```python
class student:

  def __init__(self, name , age):
    self.name = name
    self.age  = age


  def study(self):
    print(f"{self.name} is studying")


class cse_student(student):

  def __init__(self,name ,backs):
    self.name = name
    self.backs = backs


  def backlogs(self ):
    print(f"{self.name} is having {self.backs} backlogs") 
    

s1 =student("ron" , 18 )
s1.study()                  # output :  ron is studying
s2 = cse_student("tom" ,6 )
s2.study()                  # output :  tom is studying
s2.backlogs()               # output :  tom is having 6 backlogs
```
* <b> <u>Multilevel inheritance: </b></u>  When a subclass inherits properties and methods from a base class, which itself is a derived class, it is called multi-level inheritance.

```python

class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        
    def start(self):
        print("Starting the vehicle.")
        

class Car(Vehicle):
    def __init__(self, make, model, year, num_doors):
        super().__init__(make, model, year)
        self.num_doors = num_doors
        
    def drive(self):
        print("Driving the car.")
        

class ElectricCar(Car):
    def __init__(self, make, model, year, num_doors, battery_size):
        super().__init__(make, model, year, num_doors)
        self.battery_size = battery_size
        
    def charge(self):
        print("Charging the electric car.")
        

my_electric_car = ElectricCar("Tesla", "Model S", 2022, 4, 100)
my_electric_car.start()     # Output: Starting the vehicle.
my_electric_car.drive()     # Output: Driving the car.
my_electric_car.charge()    # Output: Charging the electric car.
```
* <b><u>Multiple inheritance:</b></u>  When a subclass inherits properties and methods from multiple base classes, it is called multiple inheritance.

```python

class A:
    def method_a(self):
        print("Method A")
        

class B:
    def method_b(self):
        print("Method B")

class C(A, B):
    def method_c(self):
        print("Method C")
        

my_c = C()
my_c.method_a()     # Output: Method A
my_c.method_b()     # Output: Method B
my_c.method_c()     # Output: Method C

```
* <b><u>Hierarchical Inheritance:</b></u> Several child classes inherit from the same parent class.

```python
class Parent:
    def parent_method(self):
        print("Parent Method")

class Child1(Parent):
    def child1_method(self):
        print("Child1 Method")

class Child2(Parent):
    def child2_method(self):
        print("Child2 Method")
        
```
* <b><u>Hybrid Inheritance:</b></u> This is a combination of two or more types of inheritance. For example, a class can inherit from multiple classes and also have multi-level inheritance at the same time.

```python
class Grandparent:
    def grandparent_method(self):
        print("Grandparent Method")

class FirstParent(Grandparent):
    def first_parent_method(self):
        print("First Parent Method")

class SecondParent(Grandparent):
    def second_parent_method(self):
        print("Second Parent Method")

class Child(FirstParent, SecondParent):
    def child_method(self):
        print("Child Method")

```
